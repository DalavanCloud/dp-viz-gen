# Generate visualizations used in dp-viz project

```python
import MyHeatmap
import matplotlib
%matplotlib inline

cmR9 = matplotlib.colors.ListedColormap(["#fff5f0","#fee0d2","#fcbba1","#fc9272","#fb6a4a","#ef3b2c","#cb181d","#a50f15","#67000d"])
```

**Make sure to start the notebook in the virtualenv with requirements installed.**

## Input True data (Log Scale)

```python
MyHeatmap.drawHeatMap(name = 'data/BJTaxi/in_1E6.npy',
                      transformFunc = MyHeatmap.logNorm,
                      cmap = 'RdBu_r',
                      outName = 'in_1E6',
                      dpi = 256)
```

## Direct visualization of noisy output (Log Scale)

```python
MyHeatmap.drawHeatMap(name = 'data/BJTaxi/out_Identity_1E8_e1E-1.npy',
                      transformFunc = MyHeatmap.logNorm,
                      cmap = 'RdBu_r',
                      nonNegative = True,
                      outName = 'out_Identity_NN_1E8_e1E-1',
                      dpi = 256)
```

```python
MyHeatmap.drawHeatMap(name = 'data/BJTaxi/out_Identity_1E6_e1E-1.npy',
                      transformFunc = MyHeatmap.logNorm,
                      cmap = 'RdBu_r',
                      nonNegative = True,
                      outName = 'out_Identity_1E6_e1E-1_nonNeg',
                      dpi = 256)
```

## Smoothed visualization using binned corlormap and D correction (Log Scale)

```python
MyHeatmap.drawHeatMapSmooth(name = 'data/BJTaxi/out_Identity_1E6_e1E-1.npy',
                            binNum = 9,
                            transformFunc = MyHeatmap.logNorm,
                            deltaD = 50, # remember to change when using differe eps
                            uniformBin = False,
                            cmap = cmR9,
                            outName = "out_Identity_1E6_e1E-1_smooth",
                            dpi = 256,
                            showHist = False)
```
